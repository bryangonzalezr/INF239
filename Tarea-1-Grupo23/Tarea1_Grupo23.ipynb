## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Usuarios"

```python
%%sql

CREATE TABLE Usuarios(
    ID_usuario INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR (45) NOT NULL,
    apellido VARCHAR (45),
    correo VARCHAR (45) NOT NULL,
    contraseña VARCHAR(50) NOT NULL,
    fecha_registro TIMESTAMP NOT NULL,
    PRIMARY KEY(ID_usuario)
);
```

### Tabla "Creadores"

```python
%%sql

CREATE TABLE Creadores(
    Id_creador INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    apellido VARCHAR(45) NOT NULL,
    popularidad INTEGER,
    estilo VARCHAR(45),
    PRIMARY KEY(Id_creador)
);
```

### Tabla "Tipos_NTF"

```python
%%sql

CREATE TABLE Tipos_NFT(
    ID_tipo INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(20) NOT NULL,
    PRIMARY KEY(ID_tipo)
);
```

### Tabla "NFT"

```python
%%sql

CREATE TABLE NFT(
    ID_nft INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    ID_creador INTEGER NOT NULL,
    FOREIGN KEY (ID_creador) REFERENCES Creadores(ID_creador),
    id_tipo INTEGER,
    FOREIGN KEY(ID_tipo) REFERENCES Tipos_NFT(ID_tipo),
    PRIMARY KEY(ID_nft)

);
```

### Tabla "Precio NFT"

```python
%%sql

CREATE TABLE Precio_NFT(
    ID_nft INTEGER,
    FOREIGN KEY (ID_nft) REFERENCES NFT(id_nft),
    fecha TIMESTAMP NOT NULL,
    precio INTEGER NOT NULL,
    PRIMARY KEY(id_nft, fecha)
);
```

### Tabla "Usuario tiene NFT"

```python
%%sql

CREATE TABLE Usuario_tiene_NFT(
    id_usuario INTEGER,
    id_nft INTEGER,

    FOREIGN KEY (id_usuario) REFERENCES Usuarios(id_usuario),
    FOREIGN KEY (id_nft) REFERENCES NFT(id_nft),

    PRIMARY KEY (id_usuario, id_nft)


);
```

### Consulta 7
**Top 4 NFTs más caros actualmente de un id_tipo X .**

Ejemplo:

|NFT | 
|--------|
|Minecraft Sphere | 
|The Binary Girl |
|Koalas in Love | 
|Art Dodger | 

**Solución:** Primero anidamos una consulta, donde escogemos los precios actuales de todos los NFT. Esto a través de un GROUP BY, y en el SELECT elegimos con MAX la fecha mas reciente. Luego, hacemos un INNER JOIN con el precio nuevamente (para poder obtener el precio en esa fecha), también un INNER JOIN con nft, para obtener el id_tipo correspondiente al id del nft, y hacemos un ultimo INNER JOIN con tipos_nft para obtener el tipo correspondiente. Finalmente, ponemos una condicion en la cual el nombre del tipo debe ser 'Juegos' (puede ser cambiado, este es el id_tipo X). Finalmente ordenamos por precio y lo limitamos a 4, para obtener los 4 mayores

```python
%%sql

SELECT nft.nombre AS "NFT"
FROM(
	SELECT precio_nft.id_nft,  MAX(precio_nft.fecha) AS "ultimafecha"
	FROM precio_nft
	GROUP BY precio_nft.id_nft) AS lastdate
	
	INNER JOIN precio_nft 
	ON precio_nft.fecha=lastdate.ultimafecha
	AND precio_nft.id_nft = lastdate.id_nft
	
	INNER JOIN nft
	ON precio_nft.id_nft=nft.id_nft
	
	INNER JOIN tipos_nft
	ON tipos_nft.id_tipo=nft.id_tipo
WHERE tipos_nft.nombre= 'Juegos'
ORDER BY precio_nft.precio DESC
LIMIT 4;
```

**Los precios de todos los NFTs entre los años 2017 y 2019 inclusive.**

**Creador al cual le han comprado más NFTs.**

Ejemplo:

**Solución:** Utilizamos el SELECT para que nos muestre el nombre y el apellido de "Tabla" (en un momento explicamos "Tabla") y un contador (COUNT) de la cantidad de veces que aparece un nombre en la tabla. Luego, en el FROM, lo primero que hacemos es una consulta anidada, en la cual se va a generar una nueva tabla, llamada "Tabla", la cual va a retornar el ID del creador, el ID del NFT, el nombre del creador y el apellido de este, sacando estos valores de la intersección entre la tabla NFT y la tabla creadores y agrupándola por id del creador e id del NFT. Luego, utilizamos INNER JOIN para intersectar "Tabla" con Usuario_tiene_NFT. De esta manera, vamos a obtener los datos en donde la gente haya comprado algún NFT. Agrupamos el resultado por el nombre de la tabla y el apellido, lo ordenamos desde el nombre con más cantidad de repeticiones hasta el nombre con menor cantidad, y limitamos los resultados a un solo resultado, para que nos de así el resultado esperado.

```python
%%sql

SELECT Tabla.nombre AS "Nombre", Tabla.apellido AS "Apellido", COUNT(Tabla.nombre) AS "Cantidad"
FROM (
    SELECT Creadores.ID_creador, NFT.ID_nft, Creadores.nombre, Creadores.apellido
    FROM NFT INNER JOIN Creadores ON NFT.ID_creador = Creadores.ID_creador
    GROUP BY Creadores.ID_creador, NFT.ID_nft
    ) AS Tabla
    INNER JOIN Usuario_tiene_NFT ON Tabla.id_nft = Usuario_tiene_NFT.id_nft
GROUP BY Tabla.nombre, Tabla.apellido
ORDER BY COUNT(Tabla.nombre) DESC
LIMIT 1;
```

### Consulta 10  
**El NFT con mayor precio actual, que pertenezca a un creador X.**

Ejemplo:
   
|NFT |Valor|
|-|-------|
|Minecraft Sphere |18000|

**Solución:** Comenzando de adentro hacia afuera, tenemos 2 anidaciones en total. En la mas interior, se genera una tabla-consulta con los precios solamente del año mas nuevo, esto a traves de un GROUP BY del id_nft, usando MAX para encontrar la fecha mayor(que sería la actual). Posterior a esto, hacemos un INNER JOIN entre la tabla que acabamos de crear, precio_nft, nft y los creadores. Todo esto para poder conectar la popularidad de los creadores con el precio, ya que estos no estan directamente conectados. Luego, simplemente condicionamos el nombre del creador (En este caso utilizamos 'John', pero equivale a X y puede ser cambiado por cualquiera), y luego lo ordenamos por precio para obtener el NFT que tiene un precio más alto actual perteneciente al creador X (en este caso 'John')

```python
%%sql

SELECT nft.nombre AS "NFT", precio_nft.precio AS "Valor"
 FROM
	(SELECT id_nft, MAX(fecha) AS "ultimafecha"
	 FROM precio_nft
	 GROUP BY id_nft) AS lastprice

	INNER JOIN precio_nft 
	ON lastprice.ultimafecha= precio_nft.fecha 
	AND lastprice.id_nft = precio_nft.id_nft

	INNER JOIN nft
	ON lastprice.id_nft =nft.id_nft

	INNER JOIN creadores
	ON creadores.id_creador = nft.id_creador
WHERE creadores.nombre='John'
ORDER BY precio_nft.precio DESC
LIMIT 1;
```

Además, juntamos el nombre y el apellido utilizando la sintaxis ||, en una unica columna llamada Usuario. Finalmente ordenamos en funcion de la funcion SUM en orden descendente, y limitamos a 1 para obtener la solamente al usuario que posee mayor cantidad de dinero en la cartera en función del precio actual de sus NFT.

```python
%%sql

SELECT (usuarios.nombre || ' ' || usuarios.apellido) AS "Usuario", SUM(precio_nft.precio) AS "Dinero"
FROM
	(SELECT id_nft, MAX(fecha) AS "ultimafecha"
	FROM precio_nft
	GROUP BY id_nft) AS lastprice

	INNER JOIN precio_nft 
	ON lastprice.ultimafecha= precio_nft.fecha 
	AND lastprice.id_nft = precio_nft.id_nft

	INNER JOIN usuario_tiene_nft
	ON usuario_tiene_nft.id_nft = precio_nft.id_nft
	
	INNER JOIN usuarios
	ON usuarios.id_usuario = usuario_tiene_nft.id_usuario
GROUP BY usuarios.nombre,usuarios.apellido
ORDER BY SUM(precio_nft.precio) DESC
LIMIT 1;
```
